##### 3) * Text PreProcessing

```python
df['length'] = df['text'].apply(lambda x: len(x))
df.head()
```

- Remove stopwords

```python
#Remove stopwords
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
df["preprocessed_abstract"] = df["text"].apply(lambda text: remove_stopwords(text))
df["preprocessed_abstract"] .head()
```

- Lemmatization

```python
nlp = spacy.load('en_core_web_sm')
def lemmatization(text):
    doc=nlp(text)
    lemma_list=[word.lemma_ for word in doc]
    return ' '.join(lemma_list)
df["preprocessed_abstract"] = df["preprocessed_abstract"].apply(lambda text: lemmatization(text))
df["preprocessed_abstract"] .head()
```

- Lowering and Removing punctuation

```python
nlp = spacy.blank('en')
def spacy_preprocessor(doc):
    return [t.lower_ for t in nlp(doc) if 
          not t.is_punct and t.is_alpha]

def data_preprocessing(review):
    review = ' '.join(spacy_preprocessor(review))
    return review
```

```python
df['preprocessed_abstract'] = df['preprocessed_abstract'].apply(data_preprocessing)
df['length_after_cleaning'] = df['preprocessed_abstract'].apply(lambda x: len(x))
df.head()
```

##### 4)**Visual Exploratory Analysis for column 'preprocessed_abstract'

```python
all_text = ' '.join(df['preprocessed_abstract'].astype(str).tolist())
tokens = word_tokenize(all_text)
freq_dist = FreqDist(tokens)

top_words = freq_dist.most_common(20)

words, frequencies = zip(*top_words)

plt.figure(figsize=(10, 6))
plt.bar(words, frequencies)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 20 Most Common Words')
plt.xticks(rotation=45, ha='right')  
plt.show()
```

```python
# Create a WordCloud object
wc = WordCloud(stopwords=set(list(STOPWORDS) + ['inside']), random_state=42)

# List of sentiment categories
sentiments = ['positive', 'neutral', 'negative']

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Generate and display word clouds for each sentiment category
for i, sentiment in enumerate(sentiments):
    sub_data = df[df['airline_sentiment'] == sentiment]
    op = wc.generate(' '.join(sub_data['preprocessed_abstract']))  # Assuming 'text' is the column containing the text data
    axes[i].imshow(op, interpolation='bilinear')
    axes[i].set_title(sentiment.upper(), fontsize=24)
    axes[i].axis('off')

plt.suptitle('TOP WORDS FOR A GIVEN SENTIMENT', fontsize=30)
plt.show()
```

##### 5) Preparing Data

- Tokenization

```python
df['tokenized_text'] = df['preprocessed_abstract'].apply(lambda x: word_tokenize(x))
df.head()
```

- Renaming column "airline_sentiment" to 'label' and transforming non-numerical labels (as long as they are hashable and comparable) to numerical labels.

```python
df.rename(columns={"airline_sentiment": 'label'}, inplace=True)
```

```python
label_map = {
    'negative': 0,
    'positive': 1,
    'neutral': 2
}
df['label'] = df['label'].map(label_map)
```

- Word2Vec model (pre-trained word embeddings) for embedding layer in the model

```python
from gensim.models import Word2Vec
# Create and train the Word2Vec model
model_2= Word2Vec(sentences=df['tokenized_text'], vector_size=100, window=5, min_count=1, workers=4)
# Access the word vectors
word_vectors = model_2.wv
word_vectors.vectors
```

```python
word_vectors.vectors.shape
```

```python
def get_vector_for_text(text):
    tokens = word_tokenize(text)
    vector_sum = sum([word_vectors[token] for token in tokens if token in word_vectors])
    return vector_sum / len(tokens) if len(tokens) > 0 else None
```

```python
df['word2vec_vectors'] = df['preprocessed_abstract'].apply(get_vector_for_text)
df[['preprocessed_abstract', 'word2vec_vectors']]
```
